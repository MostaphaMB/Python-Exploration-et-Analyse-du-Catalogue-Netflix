#### 2️⃣ ***Préparation et nettoyage des données***

***2.1.Gérer les valeurs manquantes***

```python
df
```

```python
df.isnull().sum()
```

```python
df["rating"].unique()
```

***2.2.chercher index des valeur manquante dans 'duration'***

```python

df.loc[df['rating']=='74 min',['rating','duration']] #i=5541
df.loc[df['rating']=='84 min',['rating','duration']]  #i=5794
df.loc[df['rating']=='66 min',['rating','duration']]  #i=5813
```

```python
# pour trouver l'index directement 
# idx=df[  (df["duration"].isna())&(df["rating"].str.contains("min",case=False)) ].index.values
# idx
```

***2.3.Remplacer les valeurs null par '74 min', '84 min', '66 min' dans la colone 'duration'***

```python

df.at[5541, 'duration'] = '74 min'
df.at[5794, 'duration'] = '84 min'
df.at[5813, 'duration'] = '66 min'
df.isnull().sum()
```

***2.4.Calculer le mode de rating***

```python
df["rating"].mode()
```

***2.5.Remplacer les valeurs incorrect par le mode 'TV-MA' dans la colone rating***

```python

df['rating']=df['rating'].replace('74 min','TV-MA')
df['rating']=df['rating'].replace('84 min','TV-MA')
df['rating']=df['rating'].replace('66 min','TV-MA')
df['rating'].unique()
```

```python
df.isnull().sum()
```

***2.6.Remplacer les valeurs null par le mode 'TV-MA' dans la colone rating***

```python

df['rating']=df['rating'].fillna('TV-MA')
df.isnull().sum()
```

***2.7.Convertir 'date_added' en Datetime***

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), format='mixed', errors='coerce')
```

```python
print(df['date_added'].dtype)
```

```python
idx=df[ (df["date_added"].isna()) ]
idx
```

***2.8.On remplit les dates vides par le 1er janvier de l'année de sortie***

```python
df['date_added'] = df['date_added'].fillna(pd.to_datetime(df['release_year'].astype(str) + '-01-01'))
```

```python
df.isnull().sum()
```

***2.9.Créer les colonnes dérivées (Année, Mois, Jour)***

```python
df['year_added'] = df['date_added'].dt.year.astype(int)
df['month_added'] = df['date_added'].dt.month.astype(int)
df['day_added'] = df['date_added'].dt.day.astype(int)
```

```python
df
```

```python
print(df['year_added'].dtype)
```

***2.10.Remplacer les autres valeurs null par 'Inconnu'***

```python
df=df.fillna('Inconnu')
df.isnull().sum()
```

***2.11.Enregister les données nettoyé***

```python
netflix_cleaned=df
netflix_cleaned.to_csv('netflix_cleaned.csv', index=False)
```

#### 3️⃣ ***Visualisations et analyse métier***

```python
df
```

***3.1. Répartition par classification d'âge (Rating)***

```python
# sns.set_theme(style="white")
plt.style.use('dark_background')
plt.figure(figsize=(12, 6))
sns.countplot(
    x='rating', 
    data=df, 
    order=df['rating'].value_counts().index,
    palette='Reds_r'
)
plt.title("Répartition des contenus par classification d'âge",fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

La catégorie TV-MA (pour public adulte) est la plus représentée, suivie par TV-14. Cela indique un positionnement clair de Netflix vers un public adulte et jeune adulte.

***3.2. Comparaison Films vs Séries***

```python
# sns.set_theme(style="whitegrid")
plt.style.use('dark_background')
plt.figure(figsize=(12, 6))
sns.countplot(
    x='year_added', 
    data=df[df['year_added'] > 2014], 
    hue='type', 
    palette=["#E50914", "#808080"]
)
# Ajout des titres
plt.title("Nombre de titres ajoutés par année sur Netflix", fontsize=15)
plt.xlabel("Année d'ajout", fontsize=12)
plt.ylabel("Nombre de films / séries", fontsize=15)

plt.show()
```

On observe un pic d'ajouts entre 2018 et 2019. Bien que le volume de films reste supérieur, le rythme de croissance des séries TV est devenu plus stable sur les dernières années.

***3.3.Nombre de contenus ajoutés chaque année*** 

```python
df_counts = df.groupby(['year_added', 'type']).size().reset_index(name='count')

plt.figure(figsize=(12, 6))
sns.lineplot(
    x='year_added', 
    y='count', 
    data=df_counts[df_counts['year_added'] > 2010], 
    hue='type', 
    style='type',
    markers=True, 
    palette=["#E50914", "#808080"]
)
plt.title("Tendances d'ajouts : Films vs Séries",fontsize=15)
plt.show()
```

La graphique en ligne montre bien le pic d'ajouts entre 2018 et 2019. Bien que le volume de films reste supérieur, le rythme de croissance des séries TV est devenu plus stable sur les dernières années.

***3.4.Proportion de films vs séries (type)***

```python
# Proportion avec (Pie Chart)
plt.figure(figsize=(8, 8))
plt.pie(
    df['type'].value_counts(), 
    labels=df['type'].value_counts().index, 
    autopct='%1.1f%%', 
    colors=["#E50914", "#808080"], 
    startangle=145
)
plt.title("Proportion de Films vs Séries",fontsize=15)
plt.show()
```

***3.5.Nombre de séries de 2011 à 2021***

```python
df_tv_shows = df[(df['type'] == 'TV Show') & (df['year_added'].between(2011, 2021))]

plt.figure(figsize=(12, 6))
sns.countplot(
    x='year_added', 
    data=df_tv_shows, 
    color="#E50914"
)
plt.title("Nombre de séries ajoutées entre 2011 et 2021",fontsize=15)
plt.show()
```

Nombre de séries ajoutées entre 2011 et 2021 montre que l'augmentation des séries a commencé en 2016 et le pic entre 2019 et 2020, puis on remarque une diminution en 2021.

***3.6.Nombre de productions par pays***

```python
top_10_countries = df['country'].value_counts().head(10).index

plt.figure(figsize=(12, 6))
sns.countplot(
    y='country', 
    data=df[df['country'].isin(top_10_countries)], 
    order=top_10_countries,
    hue='country',
    legend=False,
    palette='Reds_r'
)
plt.title("Top 10 des pays producteurs de contenu",fontsize=15)
plt.show()
```

Le graphique montre le Top Pays et les États-Unis restent le leader incontesté, suivis par l'Inde et le Royaume-Uni.

***3.7.Analyse Multi-variable : Pays et Classification (Heatmap)***

```python
top_countries = df['country'].value_counts().head(10).index
subset = df[df['country'].isin(top_countries)]
pivot_table = subset.groupby(['country', 'rating']).size().unstack().fillna(0)

plt.figure(figsize=(12, 6))
sns.heatmap(
    data=pivot_table, 
    annot=True, 
    fmt='g', 
    cmap='Reds'
)
plt.title("Nombre de productions par pays et par classification",fontsize=15)
plt.show()
```

Dans les États-Unis la catégorie TV-MA (Television Mature Audience) pour public adulte est la plus représentée , suivi de R (Restricted) utilisé pour les films (cinéma), tous deux signalant une maturité élevée, puis TV-14 (TV Parental Guidelines) catégorie déconseillé aux moins de 14 ans.
Par contre en India TV-14 represente le nombre de production le plus elevé suivi de TV-MA puis TV-PG (Parental Guidance Suggested) qui est un programme adapté aux enfants, mais qui nécessite une surveillance parentale.

***3.8.Analyse par Genres***

```python

#  On sépare les genres ET on réinitialise l'index pour éviter les doublons
genres_series = df['listed_in'].str.split(', ').explode().reset_index(drop=True)

# Conversion en DataFrame
df_genres = genres_series.to_frame(name='genre_name')

# Calcul de l'ordre pour le Top 15
top_15_order = df_genres['genre_name'].value_counts().index[:15]

plt.figure(figsize=(12, 6))

sns.countplot(
    y='genre_name', 
    data=df_genres, 
    order=top_15_order,
    # hue='genre_name',     
    palette='Reds_r',
    legend=False
)

plt.title("Top 15 des genres les plus fréquents sur Netflix", fontsize=15)
plt.xlabel("Nombre de titres")
plt.ylabel("Genres")

plt.show()
```

Le graphique montre que les genres International Movies, Dramas et Comedies forment la production essentielle de Netflix.Documentaries occupe aussi une place solide dans le Top 5, confirmant l'intérêt croissant des abonnés pour ce contenu.
